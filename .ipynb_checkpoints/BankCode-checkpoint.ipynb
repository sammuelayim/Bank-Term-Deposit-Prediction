```python
import importlib
import pip
def check_and_install_library(library_name):
    try:
        importlib.import_module(library_name)
        print(f'{library_name} has already been installed')
    except ImportError:
        try:
            pip.main(['install'],library_name)
            print(f'{library_name} has been installed')
        except:
            print("Error: Failed to install the library. Please install it manually.")
```

```python
check_and_install_library('pandas')
check_and_install_library('numpy')
check_and_install_library('matplotlib')
check_and_install_library('seaborn')
check_and_install_library('sklearn')
```

##### the day wouldn't be needed in making predictions

```python
full_df.drop('day', axis=1, inplace = True)
```

```python
dummies = pd.get_dummies(full_df, drop_first = True).astype('int')
```

```python
dummies.columns
```

###### Oversampling

```python
x = dummies.drop('y', axis=1)
y = dummies['y']

from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)
X,y = ros.fit_resample(x, y)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test= train_test_split(X,y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier(n_estimators=100, random_state=42)
model1.fit(x_train, y_train)
```

```python
y_pred1 = model1.predict(x_test)
```

```python
print(classification_report(y_test,y_pred1))
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
importances = model1.feature_importances_
feature_names = X.columns

importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

importance_df
```

```python
threshold = 0.003

important_features = importance_df[importance_df['Importance'] > threshold]['Feature']
X_filtered = X[important_features]

print("Selected features based on importance:")
print(important_features)
```

```python
X_train_filtered, X_test_filtered, y_train, y_test = train_test_split(X_filtered, y, test_size=0.2, random_state=42)


model_filtered = RandomForestClassifier(n_estimators=100, random_state=42)
model_filtered.fit(X_train_filtered, y_train)


y_pred_filtered = model_filtered.predict(X_test_filtered)

print(classification_report(y_test,y_pred_filtered))
```

```python
import joblib

# Save the model to a file
joblib.dump(model_filtered, 'xgboost_model.pkl')
```
